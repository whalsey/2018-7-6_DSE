```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

data1 = pd.read_csv("/Volumes/wms_ssd/Qualifying Exam/Input/train_2016_v2.csv")
data2 = pd.read_csv("/Volumes/wms_ssd/Qualifying Exam/Input/train_2017.csv")
```

```python
plt.close()
plt.hist(data1['logerror'], bins=100)
plt.show()
```

```python
plt.close()
plt.hist(data2['logerror'], bins=100)
plt.show()
```

```python
data = data1.append(data2)
```

```python
plt.close()
plt.hist(data['logerror'], bins=100)
plt.xlabel('logerror', fontsize=10)
plt.ylabel('count', fontsize=10)
plt.title('Logerror Distribution')
plt.show()
```

```python
plt.close()
lower = np.percentile(data['logerror'].values, 0.5)
upper = np.percentile(data['logerror'].values, 99.5)
print(data.describe([0.005, 0.995]))
print(lower)
print(upper)
plt.hist(data[(data['logerror'] > lower) & (data['logerror'] < upper)]['logerror'], bins=100)
plt.xlabel('logerror', fontsize=10)
plt.ylabel('count', fontsize=10)
plt.title('99% of Data Logerror Distribution')
plt.show()
```

```python
def crop(row):
    if row['logerror'] < -4:
        return -4
    elif row['logerror'] > 4:
        return 4
    else:
        return row['logerror']
```

```python
data1_2 = data1.copy()
data1_2['logerror'] = data1_2.apply(crop, axis=1)
```

```python
plt.close()
plt.hist(data1_2['logerror'], bins=100)
plt.show()
```
